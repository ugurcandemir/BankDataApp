```python

import pandas as pd
import numpy as np

# Load the first and second sheets of AKBNK.xlsx
sheet1 = pd.read_excel("ing_bank.xlsx", sheet_name=0)  # First sheet
sheet2 = pd.read_excel("ing_bank.xlsx", sheet_name=1)  # Second sheet

# Display the data (optional)
print(sheet1.head())
print(sheet2.head())
```

```python

# Transform the "Kalem" column into column names for the first sheet
transformed_sheet1 = sheet1.set_index("FİNANSAL DURUM TABLOSU").T.reset_index()
transformed_sheet1.rename(columns={"index": "Yıllar"}, inplace=True)
transformed_sheet1.fillna(0, inplace=True)  # Replace NaN with 0

# Transform the "Kalem" column into column names for the second sheet
transformed_sheet2 = sheet2.set_index("KAR VEYA ZARAR VE DİĞER KAPSAMLI GELİR TABLOSU").T.reset_index()
transformed_sheet2.rename(columns={"index": "Yıllar"}, inplace=True)
transformed_sheet2.fillna(0, inplace=True)  # Replace NaN with 0
```

```python
print(transformed_sheet1.head())
print(transformed_sheet2.head())
```

```python
transformed_sheet1.to_excel("ing_balance.xlsx", index=False)
transformed_sheet2.to_excel("ing_income.xlsx", index=False)
# Save the transformed data to new Excel files
```
